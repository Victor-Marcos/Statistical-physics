# Histograms

In this script we will show for a certain number of particles the histogram of their position in an specific moment of time.

```python
import randomwalk
import matplotlib.pyplot as plt
```

```python
# Parameter definition
v = 1.          # particle (average) velocity
T = 1.         # simulation duration
dt = 0.1        # elementary time step
Ndt = int(T/dt) # number of calculation steps
Npart=1000      # Number of particles
```

```python
# Initialization of the position of the current particle
title = "Number of particles as a function of its position for T= "+str(T)
# Array that will keep the position of the particles at time T
xi_arr = []
```

```python
# Loop over different particles
for ni in range(1,Npart+1):
    xi = 0
    
# Loop over all time steps, from 1 to Ndt inclusive. It goes up to T.
    for ti in range(0,Ndt+1):
    # Updates the position of the current particle
        xi = randomwalk.move_particle(xi,v,dt)

    xi_arr.append(xi) # It keeps the position of particle i at time T

# As the second element of the following list, a sufficiently large number should be entered, as this indicates the number of bins.
randomwalk.plot_histogram(xi_arr, 10000, xlabel="Position in space", ylabel="Number of particles", title=title)
plt.show()
```

```python
# Initialization of the position of the current particle
title = "Number of particles as a function of its position for T= "+str(T)
# Array that will keep the position of the particles at time T
xi_arr = []
# Loop over different particles
for ni in range(1,Npart+1):
    xi = 0
    
# Loop over all time steps, from 1 to Ndt inclusive. It goes up to T.
    for ti in range(0,Ndt+1):
    # Updates the position of the current particle
        xi = randomwalk.move_particle(xi,v,dt)

    xi_arr.append(xi) # It keeps the position of particle i at time T

# As the second element of the following list, a sufficiently large number should be entered, as this indicates the number of bins.
randomwalk.plot_histogram(xi_arr, 10000, xlabel="Position in space", ylabel="Number of particles", title=title)
plt.show()
```

In all three cases, we observe a curve resembling a Gaussian bell shape. There is a higher concentration of particles around position 0 and in its vicinity. The mean, the mode and the median are in the poition 0. However, for positions far from 0, both positive and negative, the number of particles is very low, even zero. This is because the initial position of the particles is at 0.
