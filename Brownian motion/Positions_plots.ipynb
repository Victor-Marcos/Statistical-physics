# In this script we are going to plot the position of several particles in function of time changing some parameters

```python
import numpy as np
import randomwalk
```

Let's first plot the movement of only one particle

```python
v = 1.          # average particle velocity
T = 10.         # simulation duration
dt = 0.1        # time step
Ndt = int(T/dt) # number of calculation steps
Npart=1        # number of particles we are going to plot
```

```python
title = "Position versus time, with velocity: v = " + str(v)

save_xi = []
save_ti = []
xi = 0
ti = 0
save_xi.append(xi)
save_ti.append(ti)

# Loop over all time steps, from 1 to Ndt inclusive
for ti in range(1,Ndt+1):
   
    # Updates the position of the current particle
    xi = randomwalk.move_particle(xi,v,dt)
    # Saves the current position
    save_xi.append(xi)
    # Saves the time corresponding to this position
    save_ti.append(ti)
randomwalk.plot_curve(save_ti,save_xi,xlabel="$t_i$",ylabel="$x_i$", title=title)  
```

Let's plot now 10 particles:


```python
v = 1.          # average particle velocity
T = 10.         # simulation duration
dt = 0.1        # time step
Ndt = int(T/dt) # number of calculation steps
Npart=10        # number of particles we are going to plot
```

```python
# The list save_xi will store all positions taken by the current particle
save_xi = []
# The list save_ti will store the times corresponding to these positions
save_ti = []
```

```python
title = "Position versus time, with velocity: v = " + str(v)

for ni in range(1,Npart+1):
    xi = 0
    save_xi = []
    save_ti = []

# Loop over all time steps, from 1 to Ndt inclusive
    for ti in range(1,Ndt+1):
    # Updates the position of the current particle
        xi = randomwalk.move_particle(xi,v,dt)
    # Saves the current position
        save_xi.append(xi)
    # Saves the time corresponding to this position
        save_ti.append(ti)
    randomwalk.plot_curve(save_ti,save_xi,xlabel="$t_i$",ylabel="$x_i$", title=title)  
```

It can be observed that each particle follows a different trajectory. This is because Brownian motion is a probabilistic model.

Conclusion: It can observed that if we encrease the speed of the particles, they can reach further positions in the same period of time
