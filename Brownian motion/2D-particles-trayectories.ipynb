# 2D trayectories

In this script we will show 2D particle trayectories. We will study the non-uniform probabilities too.

LetÂ´s see firest the 2D trayectory of only one particle

```python
import numpy as np
import randomwalk
import matplotlib.pyplot as plt
```

```python
# Definition of parameters
v = 1.          # Average particle speed
T = 100.         # Duration of the simulation
dt = 0.1        # Time step
Ndt = int(T/dt) # Number of calculation steps
Npart=1        # Number of particles
```

```python
# Initialization of the current particle position
title = "Particle position for T="+str(T)

for ni in range(1,Npart+1):
    xi = 0
    yi = 0
    save_xi = [] # The save_xi list will store all positions taken by the current particle in the x-axis
    save_yi = [] # The save_xi list will store all positions taken by the current particle in the y-axis
    save_ti = [] # The save_ti list will store the corresponding times for these positions

# Loop over all time steps, from 1 to Ndt inclusive
    for ti in range(1,Ndt+1):
    # Updates the position of the current particle
        xi = randomwalk.move_particle(xi,v,dt)
        yi = randomwalk.move_particle(yi,v,dt)
    # Saves the current position
        save_xi.append(xi)
        save_yi.append(yi)
    # Saves the time corresponding to this position
        save_ti.append(ti)
    randomwalk.plot_curve(save_xi,save_yi,xlabel="X",ylabel="Y", title=title)  
```

# Let's study again the mean-square position

We will create in randomwalk a function calculate_mean_squared_displacement2D that will calculate the mean squared displacement
as a function of time in 2D.

```python
# Definition of parameters
v = 1.          # Particle average speed
T = 100.         # Simulation duration
dt = 0.1        # Elementary time step
Ndt = int(T/dt) # Number of calculation steps
Npart=100       # Number of particles
prob = 0.5      # Probability that the particle moves along the X or Y axis
```

```python
# The save_xi list will store all positions taken by the current particle
save_xi = []
# The save_ti list will store the times corresponding to these positions
save_ti = []

# Initialization of the current particle position
title = "Squared Sigma as a function of time for N="+str(Npart)

#Array to store the particle positions at time T
sigma = np.zeros(Ndt)
array_temps = np.arange(0, T, dt)

#Arrays to store particle positions over time tk
xk = np.zeros(Npart)
yk = np.zeros(Npart)
# Loop over all time steps, from 1 to Ndt inclusive. It goes up to T
for tk in range(0, Ndt):
    for ni in range(0, Npart): # Loop over the number of particles   
        # Updates the position of the current particle        
        xk[ni], yk[ni] = randomwalk.move_particle2D(xk[ni], yk[ni],v,dt, prob)
    sigma[tk] = randomwalk.calculate_mean_squared_displacement2D(xk, yk)
    
randomwalk.plot_curve(array_temps, sigma,xlabel="$t_i$", ylabel="\u03C3\u00B2", title=title, linewidth=0.5)
```

# Non-uniform probabilities

We will create a code in which the particle have more chances of moving horizontally than vertically:

```python
# Definition of parameters
v = 1.          # particle's average speed
T = 100.         # simulation duration
dt = 0.1        # elementary time step
Ndt = int(T/dt) # number of calculation steps
Npart=1       # number of particles
prob = 0.99      # probability that the particle moves along the X or Y axis

# The save_xi list will save all positions taken by the current particle
save_xi = []
# The save_ti list will save the corresponding times for these positions
save_ti = []

# Initialization of the current particle's position
title = f"Particle's Position for T = {str(T)} and prob={prob}"

# Loop over different particles
for ni in range(1,Npart+1):
    xi = 0
    yi = 0
    save_xi = []
    save_yi = []
    save_ti = []
    save_xi.append(xi)
    save_yi.append(yi)
# Loop over all time steps, from 1 to Ndt inclusive
    for ti in range(0,Ndt+1):
    # Update the position of the current particle
        xi, yi = randomwalk.move_particle2D(xi, yi, v, dt, prob)
    # Save the current position
        save_xi.append(xi)
        save_yi.append(yi)
    # Save the time corresponding to this position
        save_ti.append(ti)
    randomwalk.plot_curve(save_xi ,save_yi, xlabel="$y_i$", ylabel="$x_i$", title=title) 
```

Considering a probability of 99% of moving horizontally we have obtained that there is just a few vertical movements. Let's try with probability 80%
