# Mean Square Displacement

The "mean squared displacement" refers to a statistical measure that describes the dispersion or spread of values in a dataset. It is particularly common in the analysis of random movements such as Brownian motion, where it is used to characterize the dispersion of particle trajectories over time. In the context of Brownian motion, the mean squared displacement represents the average magnitude of the square of position changes of a particle over a certain time interval. It is a useful measure for understanding the extent of particle trajectory dispersion in a given system.

```python
import numpy as np
import randomwalk
```

```python
# Parameters definition
v = 1.          # Average particle speed
T = 100.         # Simulation duration
dt = 0.1        # Elementary time step
Ndt = int(T/dt) # Number of calculation steps
Npart=100       # Number of particles
```

```python
title = "Mean square displacement as a function of time for N="+str(Npart)
#Array that will keep the particle positions at time T
sigma = np.zeros(Ndt)
array_times = np.arange(0, T, dt)

#Array that will keep the particles' positions at time ti
xk = np.zeros(Npart)

# Loop over all time steps, from 1 to Ndt inclusive. It goes up to T
for tk in range(0, Ndt):
    for ni in range(0, Npart): # Loop over the number of particles   
        # Updates the position of the current particle        
        # Met à jour la position de la particule courante
        xk[ni] = randomwalk.move_particle(xk[ni],v,dt)
    sigma[tk] = randomwalk.calculate_mean_squared_displacement(xk)

randomwalk.plot_curve(array_times, sigma,xlabel="$t_i$", ylabel="\u03C3\u00B2", title=title, linewidth=0.5)
```

We can see that the graph closely resembles a straight line. Therefore, we observe that as time progresses, the mean square displacement increases almost linearly. T
This means that as time increases, the data deviate further from the value 0, indicating a greater dispersion of the data.
